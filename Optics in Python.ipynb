# Fraunhofer diffraction

```python
import numpy as np 
import matplotlib.pyplot as plt
```

# Single Slit Diffraction

```python
b = float(input("Enter width of slit (in micrometer) : "))
l = float(input("Enter wavelength of light (in nanometer) : "))
D = float(input("Enter distance of screen to the slit (in m) : "))
#b=120
#l=632
#D=0.5
x = np.arange(-20,20,0.01)

#R is ratio of Intensity at distance x and x=0
R1 = (np.sin(np.pi*b*x/(l*D))/(np.pi*b*x/(l*D)))**2

fig, ax = plt.subplots()
ax.plot(x,R1,color = 'blue')
plt.xlabel('Fringe distance from cetre line (in mm)-->')
plt.ylabel('I(x)/I(0)-->')
ax.set_title('Single Slit diffraction pattern')
ax.grid(linestyle = ':',color='black')
plt.savefig('Single Slit Diffraction.png', dpi=300, bbox_inches='tight')
plt.show()
```
